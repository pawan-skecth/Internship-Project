
## **1. Distribution of Road Accidents Across Million-Plus Cities** **bold text**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(
    y='Million Plus Cities',
    x='Count',
    data=df,
    estimator=sum,
    ci=None,
    order=df.groupby('Million Plus Cities')['Count'].sum().sort_values(ascending=False).index
)
plt.title('Total Road Accidents by City in 2020')
plt.xlabel('Total Number of Accidents')
plt.ylabel('Million Plus Cities')
plt.tight_layout()
plt.show()
```

Distribution of Accidents Across Cities

Explanation of the Graph:

This graph shows the total number of road accident incidents reported in each
million-plus city in India in the year 2020. It uses a horizontal bar chart, where:

The Y-axis lists the names of the cities.

The X-axis shows the sum of incidents (based on the Count column).

Cities are ordered from highest to lowest total accident count.

Purpose:
This visualization helps us easily identify which cities have the highest number
of road accidents. For example, if cities like Delhi, Mumbai, or Bangalore
appear on top, it suggests they may require more focused traffic management or
road safety measures.

Insight Example:
If Delhi shows the highest count, that tells us policymakers should prioritize
road safety measures in Delhi compared to cities with fewer incidents.


# **2. Distribution of Accident Causes by Primary Cause Category**

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Cause category', data=df,
              order=df['Cause category'].value_counts().index,
              palette='magma')
plt.title('Distribution of Accident Causes by Primary Category', fontsize=16)
plt.xlabel('Number of Accidents')
plt.ylabel('Cause Category')
plt.tight_layout()
plt.show()
```


This plot helps you understand the major categories responsible for accidents,#
such as Traffic Control, Junction-related issues, Road Features, Vehicle Impact,
and Weather Conditions. It reveals which broad cause categories contribute
most to road accidents.




# **3. Detailed Analysis of Accident Causes by Subcategory**


```python
plt.figure(figsize=(12, 10))
sns.countplot(y='Cause Subcategory', data=df,
              order=df['Cause Subcategory'].value_counts().index,
              palette='coolwarm')
plt.title('Detailed Analysis of Accident Causes by Subcategory', fontsize=16)
plt.xlabel('Number of Accidents')
plt.ylabel('Cause Subcategory')
plt.tight_layout()
plt.show()
```


This graph dives deeper into the exact reasons for accidents by showing the
frequency of each specific cause under the broader categories. It highlights
which particular causes (e.g., signal violations, slippery roads) are most
common.


# **4. Outcomes of Road Accidents**


```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Outcome of Incident', data=df,
              order=df['Outcome of Incident'].value_counts().index,
              palette='Set2')
plt.title('Outcome of Road Accidents', fontsize=16)
plt.xlabel('Outcome')
plt.ylabel('Number of Incidents')
plt.tight_layout()
plt.show()
```


This bar chart shows the distribution of outcomes of road accidents, like
injuries, deaths, or other accident results. It gives insight into the severity of
accidents across cities.

## **5. Relationship Between Accident Causes and Outcomes (Stacked BarPlot)**

```python
import matplotlib.pyplot as plt

# Grouping and unstacking the data
outcome_vs_cause = df.groupby(['Cause category', 'Outcome of Incident'])['Count'].sum().unstack()

# Plotting
outcome_vs_cause.plot(
    kind='bar',
    stacked=True,
    figsize=(14, 8),
    colormap='tab20'
)
plt.title('Accident Causes vs Outcomes (Number of Incidents)', fontsize=16)
plt.xlabel('Cause Category')
plt.ylabel('Count (in Millions)')
plt.xticks(rotation=45)
plt.legend(title='Outcome of Incident', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```


This stacked bar chart shows how different accident causes are related to their
outcomes. It helps identify which causes lead to more severe outcomes like
deaths or injuries, showing the distribution of outcomes within each cause
category.

# **Bonus: Total Accident Count by City (Using the 'Count' Column)**

```python
import matplotlib.pyplot as plt

# Grouping and sorting accident counts by city
city_accident_counts = df.groupby('Million Plus Cities')['Count'].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
city_accident_counts.plot(kind='bar', color='coral')
plt.title('Total Accident Counts by City (in Millions)', fontsize=16)
plt.xlabel('City')
plt.ylabel('Accident Count (Millions)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```


Instead of just counting the number of accident records, this graph sums the
actual accident counts (in millions) by city, showing the overall accident
magnitude city-wise. This reflects the real impact better.


# **6. Top 10 Cities with Highest Accident Counts (By Count in Millions)**

```python
top_cities = df.groupby('Million Plus Cities')['Count'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='rocket')
plt.title('Top 10 Million Plus Cities with Highest Road Accident Counts (2020)',
fontsize=16)
plt.xlabel('Total Accident Count (Millions)')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

Focuses on the top 10 cities by accident magnitude. This helps prioritize areas
that might need urgent road safety interventions.


# **7. Accident Causes Breakdown Proportion (Pie Chart)**

```python
cause_counts = df['Cause category'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(cause_counts, labels=cause_counts.index, autopct='%1.1f%%',
startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Road Accident Causes by Category (2020)', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.show()
```


This pie chart visually shows the share of each cause category in all accidents,

# **8. Outcome Distribution Within Each Cause Category (Grouped BarPlot)**

```python
outcome_cause = df.groupby(['Cause category', 'Outcome of Incident'])['Count'].sum().reset_index()

plt.figure(figsize=(14,7))
sns.barplot(data=outcome_cause, x='Cause category', y='Count', hue='Outcome of Incident', palette='Set1')
plt.title('Outcome Distribution Across Different Accident Causes', fontsize=16)
plt.xlabel('Cause Category')
plt.ylabel('Accident Count (Millions)')
plt.legend(title='Outcome of Incident')
plt.tight_layout()
plt.show()
```

Shows how different outcomes (injuries, deaths, etc.) are distributed within each

# **10. Heatmap: Correlation Between Cause Category and Outcome ofIncident (Based on Counts)**

```python

cause_outcome_pivot = df.pivot_table(index='Cause category',
columns='Outcome of Incident', values='Count', aggfunc='sum').fillna(0)

plt.figure(figsize=(10,6))
sns.heatmap(cause_outcome_pivot, annot=True, fmt=".2f", cmap='YlGnBu')
plt.title('Heatmap: Accident Count by Cause Category and Outcome',
fontsize=16)
plt.xlabel('Outcome of Incident')
plt.ylabel('Cause Category')
plt.tight_layout()
plt.show()
```

The heatmap shows how the accident counts vary across combinations of cause
categories and incident outcomes, highlighting hot spots of severe outcomes.
